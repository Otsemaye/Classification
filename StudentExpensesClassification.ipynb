```python
df["Expenses"] = df[['housing', 'food', 'transportation',
       'books_supplies', 'entertainment', 'personal_care', 'technology',
       'health_wellness', 'miscellaneous']].sum(axis = 1)
```

```python
df1 = df[['age', 'gender', 'year_in_school', 'major', 'monthly_income',
       'financial_aid', 'tuition', 'preferred_payment_method', 'Expenses' ]]
df1.head()
```

```python
df1['monthly_tuition'] = round (df1['tuition']/12, 2)
df1.head()
```

```python
df1['expenses'] = df1[['Expenses', 'monthly_tuition']].sum(axis = 1)
df1.head()
```

```python
df1['total_income'] = df[["monthly_income", "financial_aid"]].sum(axis = 1)
df1.head()
```

```python
df2 = df1[['age', 'gender', 'year_in_school', 'major', 'total_income', 'preferred_payment_method', 'expenses' ]]
df2.head()
```

```python
df3 = df2
df3.head()
```

```python
df3_new = pd.get_dummies(df3, columns = ['gender', 'preferred_payment_method'])
```

```python
int_cols = df3_new.select_dtypes(include= 'bool' ).columns
int_cols
```

```python
df3_new[int_cols] = df3_new[int_cols].astype(int)

df3_new.head()
```

```python
#create a dictionary and assign the year_in_school column values

#df2['year_in_school'] = df2['year_in_school'].replace(map)
```

```python
map = {'Freshman': 1, 'Sophomore': 2, 'Junior': 3, 'Senior': 4}

df3_new['year_in_school'] = df3_new['year_in_school'].replace(map)
df3_new.head()
```

```python
df3_new['major'].value_counts()
```

```python
map1 = {'Biology':1, 'Economics':2, 'Computer Science':3, 'Engineering': 5, 'Psychology': 4}

df3_new['major'] = df3_new['major'].replace(map1)
df3_new.head()
```

```python
map2 = {"Low spenders": 0, "High spenders": 1}

df3_new['Class'] = df3_new["Class"].replace(map2)
df3_new.head()
```

## Time to train the model

```python
#Split the data into X and y
#Split your X and y into test and train
#Train the model
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
#Splitting the data

X = df3_new.drop('Class', axis = 1)
y = df3_new["Class"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state =123)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

## Evaluation

```python
# we'll be using accuarcy_score, confusion_matrix and classification_report

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, ConfusionMatrixDisplay
```

```python
acc = accuracy_score(y_test, y_pred)*100
print(f"The accuracy is {acc:.4f}%")
```
